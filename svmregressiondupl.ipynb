```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv("stockprice.csv")
df1 = df.replace(',','', regex=True)
print(df1)
```

```python
X=df1.drop(columns=['Date','Close'])
y=df1['Close'].values.astype(float)
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
SupportVectorRegModel=SVR()
SupportVectorRegModel.fit(x_train,y_train)
```

```python
y_pred=SupportVectorRegModel.predict(x_test)
print(y_pred)
```

```python
plt.plot(y_pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
```
